```python
'''Q3. What is abc module in python? Why is it used?'''
'''The abc module in Python stands for "Abstract Base Classes." It provides infrastructure for defining abstract base classes in Python. Abstract base classes are classes that cannot be
instantiated directly but serve as a blueprint for other classes to inherit from. They define a common interface and provide a way to enforce a specific structure or behavior for their
subclasses.'''
'''Defining Abstract Base Classes (ABCs): The abc module provides the ABC class, which is used as a metaclass for defining abstract base classes. By subclassing ABC, a class becomes an
abstract base class, indicating that it is intended to be used as a base class and cannot be instantiated directly. Abstract base classes typically define one or more abstract methods
that subclasses must implement.'''

```

```python
'''Q4. How can we achieve data abstraction?'''
'''Data abstraction in Python can be achieved through the use of classes and objects. Here are the steps to achieve data abstraction:

Define a Class: Start by defining a class that represents the abstraction you want to create. This class should encapsulate the data and behavior related to the abstraction. The internal
details of how the data is stored and processed should be hidden from external entities.

Hide Internal Data: Encapsulate the data within the class by using private data members. In Python, private data members are conventionally marked by prefixing them with a single
underscore (_). This indicates that they should not be accessed directly from outside the class.

Provide Public Methods: Define public methods (also known as accessor and mutator methods) to interact with the data. These methods should provide controlled access to the internal 
data and ensure data integrity. Public methods can be used to retrieve or modify the encapsulated data.

Access Data Through Methods: Access the data through the public methods provided by the class, rather than directly accessing the private data members. This ensures that the data 
is accessed and modified using the defined interfaces, enforcing encapsulation and data hiding.'''
```

```python
'''Q5. Can we create an instance of an abstract class? Explain your answer.'''
'''No, we cannot create an instance of an abstract class in Python. An abstract class is a class that is meant to be inherited from and serves as a blueprint for its subclasses. It 
contains one or more abstract methods that do not have an implementation in the abstract class itself.'''
'''The purpose of an abstract class is to provide a common interface or contract for its subclasses to follow. It defines the structure and behavior that the subclasses must implement.
As a result, an abstract class cannot be instantiated directly because it is incomplete and lacks the implementation for the abstract methods.'''

```
